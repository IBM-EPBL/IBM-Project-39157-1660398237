# <font color="black"> 6.Handling outliers </font>

```python
#outlier Removal
```

```python
#1. using percentile

    # Here outliers are considered as the extreme values of the dataframe ( values above 99% and values below 1%)
```

```python
#Identifying the maximum value in he dataframe
```

```python
max_threshold = df["EstimatedSalary"].quantile(0.99)
max_threshold
```

```python
 new_df = df[df["EstimatedSalary"]>max_threshold]
new_df.shape
```

```python
#Identifying the minimum value in he dataframe
```

```python
min_threshold = df["EstimatedSalary"].quantile(0.01)
min_threshold
```

```python
new_df = df[df.EstimatedSalary < min_threshold]
new_df.shape
```

```python
# After removing the outliers
```

```python
df = df[(df.EstimatedSalary > min_threshold) & (df.EstimatedSalary < max_threshold)]
df.shape
```

```python
#2.using standard deviation

 #Here the values ranging between +3 std and -3std are considered as outliers
```

```python
plt.hist(df.CreditScore , bins=50 ,rwidth=0.8)
plt.xlabel("Credit Score")
plt.ylabel("frequency")
plt.show()
```

```python
min = df.CreditScore.min()
min
```

```python
max = df.CreditScore.max()
max
```

```python
#Drawing a bell curve to depict the normal distribution
```

```python
from scipy.stats import norm

plt.hist(df.CreditScore , bins=50 , rwidth=0.8 ,density = True)
plt.xlabel("Credit Score")
plt.ylabel("frequency")

rng = np.arange(min,max,0.1)
plt.plot(rng,norm.pdf(rng , df.CreditScore.mean(),df.CreditScore.std()))
```

```python
#mean
```

```python

mean = df.CreditScore.mean()
mean
```

```python
#standard Deviation
```

```python

std = df.CreditScore.std()
std
```

```python
upper_limit = mean + 3*std
upper_limit
```

```python
lower_limit = mean - 3*std
lower_limit
```

```python
# Spotting out  the outliers
```

```python
outliers = df[(df.CreditScore > upper_limit) | (df.CreditScore< lower_limit)]

outliers
```

```python
outliers.shape
```

```python
df.drop(["RowNumber","Surname","CustomerId"] , axis =1 ,inplace=True)
```

```python
df.shape  #After removing the unneccesary columns for analysis
```

```python
# Removing the outliers
```

```python
df = df[(df.CreditScore < upper_limit) & (df.CreditScore > lower_limit)]
df.shape
```

```python
df.CreditScore.value_counts()
```
