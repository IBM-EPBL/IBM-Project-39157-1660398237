## 9. Generate the series of dates from 1st Jan, 2023 to 10th Feb, 2023

```python
import pandas as pd
  
dRan1 = pd.date_range(start ='1-1-2023',
           end ='2-10-2023', freq ='D')

print(dRan1, '\n')
```
